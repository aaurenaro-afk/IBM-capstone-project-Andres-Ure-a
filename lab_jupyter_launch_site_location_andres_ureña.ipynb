```python
# TASK 3: Calculate the distances between a launch site to its proximities

import math

# Function to calculate distance between two coordinates using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians 
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a)) 
    r = 6371  # Radius of earth in kilometers
    return c * r
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create markers and lines for all proximities
for site_name, data in proximity_data.items():
    launch_site_coords = data['launch_site']
    
    # Add launch site marker
    folium.Marker(
        launch_site_coords,
        popup=f'<b>{site_name}</b><br>Launch Site',
        icon=folium.Icon(color='black', icon='rocket', prefix='fa')
    ).add_to(site_map)
    
    # Process each proximity type
    for proximity_type, proximity_info in data.items():
        if proximity_type != 'launch_site':  # Skip the launch site itself
            proximity_coords = proximity_info['coords']
            proximity_name = proximity_info['name']
            
            # Calculate distance
            distance = calculate_distance(
                launch_site_coords[0], launch_site_coords[1],
                proximity_coords[0], proximity_coords[1]
            )
            
            # Create proximity marker with distance
            folium.Marker(
                proximity_coords,
                popup=f'''
                <b>{proximity_name}</b><br>
                Type: {proximity_type.title()}<br>
                Distance to {site_name}: <b>{distance:.2f} KM</b>
                ''',
                icon=folium.Icon(
                    color=proximity_colors[proximity_type],
                    icon=proximity_icons[proximity_type],
                    prefix='fa'
                ),
                tooltip=f'{proximity_name} - {distance:.2f} KM'
            ).add_to(site_map)
            
            # Draw line between launch site and proximity
            line = folium.PolyLine(
                locations=[launch_site_coords, proximity_coords],
                weight=2,
                color=proximity_colors[proximity_type],
                opacity=0.7,
                popup=f'''
                <b>{site_name} to {proximity_name}</b><br>
                Distance: <b>{distance:.2f} KM</b><br>
                Type: {proximity_type.title()}
                ''',
                tooltip=f'Click for distance info'
            )
            site_map.add_child(line)
            
            # Add distance label at midpoint
            mid_lat = (launch_site_coords[0] + proximity_coords[0]) / 2
            mid_lon = (launch_site_coords[1] + proximity_coords[1]) / 2
            
            folium.Marker(
                [mid_lat, mid_lon],
                icon=folium.DivIcon(
                    html=f'''
                    <div style="font-size: 9pt; color: white; background: {proximity_colors[proximity_type]}; 
                           padding: 3px; border-radius: 4px; border: 1px solid white;">
                        <b>{distance:.1f} KM</b>
                    </div>
                    '''
                ),
                popup=f'Distance: {distance:.2f} KM'
            ).add_to(site_map)

# Display the map
site_map
```
